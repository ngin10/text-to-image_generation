```python
from diffusers import StableDiffusionPipeline
import torch
model_id1="dreamlike-art/dreamlike-diffusion-1.0"
modek_2="stabilityai/stable-diffusion-xl-base-1.0"
import matplotlib.pyplot as plt
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)  # Remove variant="fp16"
pipe.to("cuda")
```

```python
prompt="create a indian girl with sharp jawline and long neon coloured hairs as a musician"
```

```python
image=pipe(prompt).images[0]
```
